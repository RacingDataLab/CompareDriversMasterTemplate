```python
fastf1.Cache.enable_cache('Cache')
```

```python
year1= 2023
gp1 = 'Bahrain'
event1 = 'Q'
session_race1 = fastf1.get_session(year1, gp1, event1)
session_race1.load()
LapDrive1 = 10
driver1 ='LEC'
driver1_laps = session_race1.laps.pick_driver(driver1)
driver1_specific_lap = driver1_laps[driver1_laps['LapNumber']==LapDrive1].iloc[0]
driver1_car = driver1_specific_lap.get_car_data()
```

```python
year2= 2022
gp2 = 'Bahrain'
event2 = 'Q'
session_race2 = fastf1.get_session(year2, gp2, event2)
session_race2.load()
LapDrive2 = 11
driver2 ='LEC'
driver2_laps = session_race2.laps.pick_driver(driver2)
driver2_specific_lap = driver2_laps[driver2_laps['LapNumber']==LapDrive2].iloc[0]
driver2_car = driver2_specific_lap.get_car_data()
```

```python
plotting.setup_mpl()

fig, ax = plt.subplots(5,figsize=(14,10))

color_driver1 = "red"
color_driver2 = "blue"

ax[0].plot(driver2_car['Time'], driver2_car['Speed'], color=color_driver2)
ax[0].plot(driver1_car['Time'], driver1_car['Speed'], color=color_driver1)

ax[1].plot(driver2_car['Time'], driver2_car['RPM'], color=color_driver2)
ax[1].plot(driver1_car['Time'], driver1_car['RPM'], color=color_driver1)

ax[2].plot(driver2_car['Time'], driver2_car['nGear'], color=color_driver2)
ax[2].plot(driver1_car['Time'], driver1_car['nGear'], color=color_driver1)

ax[3].plot(driver2_car['Time'], driver2_car['Throttle'], color=color_driver2)
ax[3].plot(driver1_car['Time'], driver1_car['Throttle'], color=color_driver1)

ax[4].plot(driver2_car['Time'], driver2_car['Brake'], color=color_driver2)
ax[4].plot(driver1_car['Time'], driver1_car['Brake'], color=color_driver1)

ax[0].set_ylabel("Speed [km/h]")
ax[1].set_ylabel("RPM [#]")
ax[2].set_ylabel("Gear [#]")
ax[3].set_ylabel("Throttle [%]")
ax[4].set_ylabel("Brake [%]")

ax[0].get_xaxis().set_ticklabels([])
ax[1].get_xaxis().set_ticklabels([])
ax[2].get_xaxis().set_ticklabels([])
ax[3].get_xaxis().set_ticklabels([])
ax[4].get_xaxis().set_ticklabels([])

fig.align_ylabels()

fig.legend([driver1+" GP: "+gp1+" Year: "+str(year1)+" Lap Number: "+str(LapDrive1),driver2+" GP: "+gp2+" Year: "+str(year2)+" Lap Number: "+str(LapDrive2)])

plt.subplots_adjust(left=0.06 ,right=0.99, top=0.9, bottom=0.05)
plt.show()
```
